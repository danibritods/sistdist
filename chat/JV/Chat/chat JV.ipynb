```python
from xmlrpc.server import SimpleXMLRPCServer
import random

class Server:

    def get_db ():
        '''returns a dictionary simulating a database'''
        db_agenda = {
            'agda': '(11) 9 9911-1111',
            'beto': '(22) 9 9922-2222',
            'ciro': '(33) 9 9933-3333',
            'deni': '(44) 9 9944-4444'
        }
        return db_agenda

    def agenda(name, self):
        '''Programa base'''
        agenda_db = self.get_db()
        if name in agenda_db:
            return agenda_db[name]
        return 'not found'

    def _set_db(self):
        '''asd'''
        self.db = self.get_db()

    def _set_address(self, ip, port):
        '''asd'''
        self.ip = ip
        self.port = port
        message = 'Using ' + self.ip + ':' + self.port
        print(message)

    def _set_id(self):
        '''asd'''
        self.id = random.randint(1000, 9999)
        message = 'Creating new server object with id:' + self.id

    def _configure_server(self):
        '''asd'''
        messages = {
            'first_message': '',
            'await_message': ''
        }
        self.server = SimpleXMLRPCServer((self.ip, int(self.port)))
        
        message = self.ip + ':' + self.port
        message += 'SERVER CONNECTED AT ' + message
        self.server.register_function(self.agenda, 'agenda')
        self.server.register_function(self.get_db, 'get_db')

    def start_server(self):
        '''asd'''
        print('Waiting for clients . . .')
        self.server.serve_forever()

    def __init__(self, ip='127.0.0.1', port='8080'):
        '''Creating attributes'''
        self._set_id(self)
        self._set_db(self)
        self._set_address(self, ip, port)
        self._configure_server(self)
```

```python
import xmlrpc.client
import random

class Client:

    def _set_address(self, ip, port):
        '''asd'''
        self.ip = ip
        self.port = port
        message = 'Using ' + self.ip + ':' + self.port
        print(message)

    def _set_id(self):
        '''asd'''
        self.id = random.randint(1000, 9999)
        message = 'Creating new server object with id:' + self.id
        print(message)

    def _configure_server(self):
        '''asd'''
        first_message = 'Establishing server connection to '
        address = self.ip+':'+self.port
        server_url = 'http://'+address+'/'
        message = first_message + server_url
        print(message)
        self.server = xmlrpc.client.ServerProxy(server_url)

    def _fill_with_spaces (str_name, size_to_reach):
        '''aesthetics'''
        return ' '*(size_to_reach-len(str_name))

    def _append_borders (str_name):
        '''aesthetics'''
        return '| ' + str_name + ' |\n'

    def _contacts_capsule (contact_list, self):
        '''Just prints the capsule with the messages'''
        header = 'Contact names'
        all_lines = []
        all_lines.append(header)
        all_lines.append(contact_list)
        max_size = max(list(map(len, all_lines)))
        line = '-'*(max_size)
        ab_line = self._append_borders(line)
        ab_header = self._append_borders(header)
        message = ab_line + ab_header + ab_line

        for elem in contact_list:
            message += self._append_borders(elem + self._fill_with_spaces(elem, max_size))
        message += ab_line + '\n'

        print(message)

    def _contact_capsule (name, phone_number, self):
        '''just a pretty box for the found phone number'''
        header = 'Phone of ' + name.upper()
        lines = [header, phone_number]
        max_size = max(list(map(len, lines)))
        line = '-'*max_size
        filled_header = header + self.fill_with_spaces(header, max_size)
        ab_header = self._append_borders(filled_header)
        filled_phone_number = phone_number + self.fill_with_spaces(phone_number, max_size)
        ab_phone_number = self._append_borders(filled_phone_number)

        ab_line = self._append_borders(line)
        message =  ab_line + ab_header
        message += ab_line + ab_phone_number
        message += ab_line + '\n'
        print(message)

    def contact_server(self):
        '''doing stuff with the server'''
        name = ''
        messages = {
            'input_message': "\n Type:\n",
            'options': "\t - A person's name\n\
        \t - 'get names' to see all names\n\
        \t - 'exit' to quit the program\n",
            'error_message': '\n No match for this name',
            'exit_message': '\n Exiting program \n',
            'line': 18*'= '
        }

        while name != 'exit':
            name = ''
            name = input(messages['input_message'] + messages['options']).lower()

            if name == 'get names':
                self._contacts_capsule(list(self.server.get_db().keys()))
                continue
            if name == 'exit':
                print(messages['exit_message'])
                continue

            phone_number = self.server.agenda(name)
            if phone_number == 'not found':
                print(messages['error_message'])
            else:
                self._contact_capsule(name, phone_number)
            print (messages['line'])

    def __init__(self, ip='127.0.0.1', port='8080'):
        '''Creating attributes'''
        self._set_id(self)
        self._set_address(self, ip, port)
        self._configure_server(self)
```
